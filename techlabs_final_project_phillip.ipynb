### Constants

```
csv_url_cases = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
csv_url_deaths = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"
csv_url_recovered = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv"
```

### Usefull databases

this is the DB that appear in kaggle.

```
df_cases = pd.read_csv(csv_url_cases)
df_cases.head()
```

```
df_deaths = pd.read_csv(csv_url_recovered)
df_deaths.head()
```

```
df_recovered = pd.read_csv(csv_url_recovered)
df_recovered.head()
```

```
df_cases.drop(columns = ["Lat", "Long", "Province/State"], inplace = True)
df_deaths.drop(columns = ["Lat", "Long", "Province/State"], inplace = True)
df_recovered.drop(columns = ["Lat", "Long", "Province/State"], inplace = True)
```

```
df_cases.set_index("Country/Region", inplace=True)
df_deaths.set_index("Country/Region", inplace=True)
df_recovered.set_index("Country/Region", inplace=True)
```

```
df_cases = df_cases.groupby(df_cases.columns, axis=1).sum()
```

```
df_cases = df_cases.sum(axis = 1, level = 0)
```

```
df_cases.loc["Canada"]
```

```
df_cases["type"] = "cases"
df_deaths["type"] = "deaths"
df_recovered["type"] = "recovered"
```

```
df_cases_deaths = df_cases.merge(df_deaths, left_on = df_cases.index, right_on = df_deaths.index, how = "inner")
```
